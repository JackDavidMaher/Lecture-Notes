If $\frac{1}{\Omega} e_i \times e_j = \epsilon_{ijk} e^*_k$ where $\Omega = e_1 \cdot (e_2 \times e_3)$, what is $e^*_m$ equal to?

$$ \frac{1}{2\Omega} \epsilon_{mij} (e_i \times e_j) = \frac{1}{2} \epsilon_{mij} \epsilon_{ijk} e^*_k$$
$$\frac{1}{2\Omega} \epsilon_{mij} (e_i \times e_j) = e^*_m$$

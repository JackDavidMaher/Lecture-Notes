If $\frac{1}{\Omega} e_i \times e_j = \epsilon_{ijk} e^*_k$ where $\Omega = e_1 \cdot (e_2 \times e_3)$, what is $e^*_m$ equal to?

$$ \frac{1}{2\Omega} \epsilon_{mij} (e_i \times e_j) = \frac{1}{2} \epsilon_{mij} \epsilon_{ijk} e^*_k$$
$$\frac{1}{2\Omega} \epsilon_{mij} (e_i \times e_j) = e^*_m$$

$ \Lambda^{\mu \nu}=x^\mu p^\nu-x^\nu p^\mu$ and $L=r\times p$
$$L_i=(r\times p)_i=\epsilon_{ijk}x^jp^k.$$
$$L_i=\frac{1}{2}\epsilon_{ijk}x^jp^k+\frac{1}{2}\epsilon_{ijk}x^jp^k.$$
$$\frac{1}{2}\epsilon_{ijk}x^jp^k\rightarrow\frac{1}{2}\epsilon_{ikj}x^kp^j=-\frac{1}{2}\epsilon_{ijk}x^kp^j,$$
$$L_i=\frac{1}{2}\epsilon_{ijk}x^jp^k-\frac{1}{2}\epsilon_{ijk}x^kp^j=\frac{1}{2}\epsilon_{ijk}(x^jp^k-x^kp^j).$$
Using $\Lambda^{jk}=x^jp^k-x^kp^j$, we get
$$L_i=\frac{1}{2}\epsilon_{ijk}\Lambda^{jk}$$

Let $A^\mu=(0,\mathbf{a} \;\rm{e}^{ i (\mathbf{k}.\mathbf{x}-\omega t)})$ where a is a constant vector and $F^{\mu \nu}=\nabla^\mu A^\nu-\nabla^\nu A^\mu$ where $A^\mu$ obeys $\square A^\mu=0$. What is $F^{0i}$?

Since $A^0=0$, we have
$$F^{0i}=\nabla^0 A^i-\nabla^i A^0=\nabla^0 A^i.$$
Now
$$A^i=a^i\,\mathrm{e}^{i(\mathbf{k}\cdot\mathbf{x}-\omega t)}.$$
Differentiating with respect to the time component gives
$$F^{0i}=\frac{\partial}{\partial t}\left(a^i\,\mathrm{e}^{i(\mathbf{k}\cdot\mathbf{x}-\omega t)}\right)=-i\omega a^i\,\mathrm{e}^{i(\mathbf{k}\cdot\mathbf{x}-\omega t)}.$$
So, in the convention used here,
$$F^{0i}=-i\omega a^i\,\mathrm{e}^{i(\mathbf{k}\cdot\mathbf{x}-\omega t)}.$$
If instead you are using $x^0=ct$, this is the same result up to the usual factor of $1/c$:
$$F^{0i}=-\frac{i\omega}{c}a^i\,\mathrm{e}^{i(\mathbf{k}\cdot\mathbf{x}-\omega t)}.$$
The condition $\square A^\mu=0$ tells you this is a free wave, so $\omega=c|\mathbf{k}|$ (or $\omega=|\mathbf{k}|$ in units with $c=1$), but that is not needed to compute $F^{0i}$.

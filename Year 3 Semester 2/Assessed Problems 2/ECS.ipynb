$$ \delta(z=10) = \delta(z=1100) \frac{1101}{11} $$
Writing $\delta(z=1100) = \delta_{\text{dec}}$, above is approximated to $$ \delta(z=10) \approx 100 \delta_{\text{dec}} $$
If $$\delta(z=10) \gtrsim 1$$ then $$\delta_{\text{dec}} \gtrsim 0.01$$

$$ r_s = \frac{R_\text{vir}}{3} $$

At $r=0.05R_\text{vir}$,
$$ r/r_s = \frac{0.05R_\text{vir}}{R_\text{vir}/3} = 0.15 $$
At $r=0.75R_\text{vir}$,
$$ r/r_s = \frac{0.75R_\text{vir}}{R_\text{vir}/3} = 2.25 $$
Therefore (using the python script to calculate) the cooling time at $r=0.05R_\text{vir}$ is $1.25 \times 10^{15} \text{s}$ and the cooling time at $r=0.75R_\text{vir}$ is $1.5 \times 10^{17} \text{s}$.

The inner halo cools more efficiently than the outer halo as the cooling time is shorter. This is probably because the inner halo is denser than the outer halo and so there are more particles to collide and radiate away energy.



#### 4)
As the cooling is short in the center of the cluster, we would expect there to be a lot of star formation. As this isnt the case there must be feedback that is causing a heating effect on the gas.

```python
import numpy as np
import scipy.constants as const

def tCool(n):
    t = (3 * const.k * 10**(7)) / (2 * n * 0.8 *10**(-36))
    return t

def densityg(rRvirRatio):
    rhom = (4 * 6 * 10**(4))/((rRvirRatio)*(1 + rRvirRatio)**2)
    rhog = rhom * 0.17
    return rhog
print(f'The cooling time at r = 0.05 R_vir is {np.format_float_scientific(tCool(densityg(0.15)))}')
print(f'The cooling time at r = 0.75 R_vir is {np.format_float_scientific(tCool(densityg(2.25)))}')
```

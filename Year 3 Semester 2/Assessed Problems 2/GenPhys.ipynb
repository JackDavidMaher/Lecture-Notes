# Wedge reflection diagram
The plot below visualises the air wedge between two glass plates and labels the two reflected rays discussed in the solution: 
- `R_A`: reflection from the lower surface of the top plate (glass→air) — no π phase flip.
- `R_B`: reflection from the upper surface of the bottom plate (air→glass) — π phase flip.

```python
import matplotlib.pyplot as plt
import numpy as np
# Geometry parameters (adjustable)
L = 4.0  # wedge length (horizontal)
d = 0.6  # separation at far edge
# choose a point of incidence x0 where light hits the wedge region
x0 = 2.5
# local thickness at x0
t = (d / L) * x0
# coordinates for the two inner surfaces of the plates
xs = np.linspace(0, L, 200)
y_top = np.zeros_like(xs)
y_bottom = - (d / L) * xs
fig, ax = plt.subplots(figsize=(7,5))
# draw inner surfaces (air gap)
ax.plot(xs, y_top, color='saddlebrown', linewidth=6, solid_capstyle='butt')
ax.plot(xs, y_bottom, color='saddlebrown', linewidth=6, solid_capstyle='butt')
# draw outer edges lightly to suggest plates
ax.plot([0,L],[0, -d], color='k', linewidth=1, linestyle='--', alpha=0.4)
# incident ray from upper-left to (x0, 0)
yi = 2.0
ax.plot([x0-0.6, x0], [yi, 0.0], color='tab:blue', lw=2)
ax.text(x0-0.7, yi-0.05, 'Incident ray', color='tab:blue')
# reflected ray R_A: reflected back to upper-left (separated)
ax.plot([x0, x0-0.9], [0.0, yi], color='tab:green', lw=2)
ax.text(x0-1.05, yi+0.02, 'R_A\n(no π flip)', color='tab:green')
# R_B: transmitted into wedge, reflected at bottom surface, emerges to upper-right
ax.plot([x0, x0+0.12], [0.0, -t], color='tab:orange', lw=2)
ax.plot([x0+0.12, x0+0.35], [-t, yi], color='tab:orange', lw=2)
ax.text(x0+0.36, yi-0.1, 'R_B\n(π flip)', color='tab:orange')
# annotate the round-trip path 2t inside the wedge
ax.annotate('', xy=(x0-0.2, 0.0), xytext=(x0-0.2, -t), arrowprops=dict(arrowstyle='<->'))
ax.text(x0-0.55, -t/2.0, '2t', fontsize=12)
# labels and limits
ax.set_xlim(-0.5, L+0.8)
ax.set_ylim(-d-0.6, yi+0.4)
ax.set_aspect('equal')
ax.axis('off')
plt.title('Air wedge: inner surfaces and separated reflected rays (R_A and R_B)')
plt.show()
```

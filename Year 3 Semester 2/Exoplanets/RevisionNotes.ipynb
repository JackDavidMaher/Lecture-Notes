$$ g(r) = \frac{GM_p}{r^2} $$
$$ \frac{dg}{dr} = -\frac{2GM_p}{r^3} $$
$$ \Delta g_{\text{tidal}} \approx \left| \frac{dg}{dr} \right|_{r=a} R_s = \frac{2GM_p R_s}{a^3} $$
$$ g_s = \frac{GM_s}{R_s^2} $$
$$ M_s = \frac{4}{3} \pi R_s^3 \rho_s $$
$$ g_s = \frac{G(\frac{4}{3} \pi R_s^3 \rho_s)}{R_s^2} = \frac{4}{3} \pi G R_s \rho_s $$
At the Roche limit, the tidal acceleration equals the surface gravity:
$$ \frac{2GM_p R_s}{a^3} \approx \frac{4}{3} \pi G R_s \rho_s $$
$$ \frac{M_p}{a^3} \approx \frac{2 \pi}{3} \rho_s $$
$$ a^3 \approx \frac{3M_p}{2 \pi \rho_s} = 2R^3_p \frac{\rho_p}{\rho_s} $$
$$ a \approx R_p \left( 2 \frac{\rho_p}{\rho_s} \right)^{1/3} $$
$$ a \approx R_p \left( 2 \frac{\rho_p}{2 \rho_s} \right)^{1/3} $$
$$ a \approx 2^{-1/3} R_p \approx 0.794 R_p $$

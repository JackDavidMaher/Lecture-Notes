Protoplanetary Disks and Planet Formation

As seen on slide 84, a protoplanetary disk profile has thickness z on either side of the midplane where r is the radial distance of the protoplanet from the star and R is the distance along the midplane from the star.

### Mass density
rho(z) = mu m_p+ n(0)e^(-E(z)/kT)
E(z) = 1/2 (G mu m_p+ M_star / r^3) * z^2

scale height H = sqrt(kT R^3 / G M_star mu m_p+)

Equilibrium temperature T_p = T_star * (1-A)^1/4 * sqrt(R_star / 2a)

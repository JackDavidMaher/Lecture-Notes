### The Two Body Problem
$$ \bold{F}_1 = m \ddot{\bold{r}}_1 = \frac{Gm_1m_2}{r^3} \bold{r}$$
$$ \bold{F}_2 = m \ddot{\bold{r}}_2 = -\frac{Gm_1m_2}{r^3} \bold{r}$$
Where $\bold{r} = \bold{r}_2 - \bold{r}_1$ and $\ddot{\bold{r}}=\ddot{\bold{r_2}}-\ddot{\bold{r_1}}$
$$\ddot{\bold{r}} - \frac{G m_1 \bold{r}}{r^3} + \frac{G m_2 \bold{r}}{r^3} = 0 $$
$$\ddot{\bold{r}} + G (m_1 + m_2) \frac{\bold{r}}{r^3}  = 0 $$
$$\ddot{\bold{r}} +\mu \frac{\bold{r}}{r^3} = 0$$

Taking the cross product of r with the two body equation gives:
$$ \bold{r} \times \ddot{\bold{r}} = -\mu \frac{\bold{r} \times \bold{r}}{r^3} = 0 $$
So angular momentum is conserved.
$$ \bold{r} \times \dot{\bold{r}} = \bold{h} $$
$$\bold{h} = \begin{vmatrix} \bold{r} \\\ 0 \\\ 0 \end{vmatrix} \times \begin{vmatrix} \dot{r} \\\ r \dot{\theta} \\\ 0 \end{vmatrix} = \begin{vmatrix} 0 \\\ 0 \\\ r^2 \dot{\theta} \end{vmatrix} = r^2 \dot{\theta} \hat{z}$$
by substitution of $r=u^{-1}$ we get:
$$ \frac{d^2 u}{d \theta^2} + u = \frac{\mu}{h^2} $$
The solution to this differential equation is:
$$ u(\theta) = \frac{\mu}{h^2} [1 + e \cos(\theta - \theta_0)] $$
Where e is the eccentricity and $\theta_0$ is the angle of periapsis.
#### Kepler's First Law
$$ r(\theta) = \frac{h^2 / \mu}{1 + e \cos(\theta - \theta_0)} $$
This describes a conic section with one focus at the origin.
#### Kepler's Second Law
$$ h = \sqrt{\mu a (1-e^2)} $$
$$ r = \frac{a(1-e^2)}{1 + e \cos(f)} $$
$$ dA = \int_{0}^{f} r dr d\theta $$
$$ \dot{A} = \frac{h}{2} = \frac{1}{2} \sqrt{\mu a (1-e^2)} = \text{constant} $$
#### Kepler's Third Law
$$ dA = \frac{h}{2} dT $$
$$ A = \frac{h}{2} T $$
$$ T = \frac{2A}{h} = \frac{2 \pi a^2 \sqrt{1-e^2}}{\sqrt{\mu a (1-e^2)}} $$
$$ T^2 = \frac{4 \pi^2}{\mu} a^3 = \frac{4 \pi^2}{G (m_1 + m_2)} a^3 $$
#### Vis-Viva Equation
$$ \frac{1}{2} v^2 - \frac{G (m_1 + m_2)}{a} = -\frac{G (m_1 + m_2)}{2a} $$
$$ v^2 = \frac{GM_1}{a} \frac{1+e}{1-e} $$

# Detection Methods
### Radial Velocity Method
#### The Doppler Effect
$$ \Delta \lambda = \lambda_{obs} - \lambda_{0} $$
$$ \lambda_{obs} = \lambda_{0} \frac{1 + v/c}{\sqrt{1 - \frac{v^2}{c^2}}} $$
which for $v \ll c$ reduces to:
$$ v=\frac{\Delta \lambda}{\lambda_0} c =V_{rad} = v \sin i $$
Where i is the inclination of the planet's orbit to the line of sight (i=90 degrees is edge on and i=0 degrees is face on).

#### Moving Around the Barycenter
$$ a = a_1 + a_2 $$
Where a is the distance between the star and the planet, a1 is the distance from the star to the barycenter and a2 is the distance from the planet to the barycenter.
$$ T_1 = T_2 = T $$
$$ e_1 = e_2 = e $$
$$ a_1 : a_2 : a = m_2 : m_1 : M $$
Where $ (M = m_1 + m_2) $
Kepler's Third Law gives:
$$ \frac{4 \pi^2}{G T^2} = \frac{M}{a^3} = \frac{m_2^3 / M^2}{a_1^3} = \frac{m_1^3 / M^2}{a_2^3} $$

#### Microlensing
Light deflection angle: $ \epsilon = \frac{4GM_L}{r c^2} $

Lens Equation: $ \theta_1^2 - \alpha \theta_1 - \theta_E^2 = 0 $

With Einstein angle: $ \theta_E = \sqrt{\frac{4GM_L}{\mu D_L c^2}} = \sqrt{\frac{2R_{\text{Sch}}}{\mu D_L}}$ ($\mu = 1+\frac{D_{L}}{D_SL}$)

$ u = \frac{\alpha}{\theta_E} $
$$ \frac{\theta_1^2}{\theta_E^2} - \frac{\alpha \theta_1}{\theta_E^2} - \frac{\theta_E^2}{\theta_E^2} = 0 $$

$y = \frac{\theta_1}{\theta_E}$
$$ y^2 - u y - 1 = 0 $$
$$ y = \frac{u \pm \sqrt{u^2 + 4}}{2} $$
$$ \theta_\pm = \frac{\theta_E}{2} (u \pm \sqrt{u^2 + 4}) $$
Magnification: $ A = \frac{u^2 + 2}{u \sqrt{u^2 + 4}} $

The timescale, $ t_E = \frac{\theta_E}{\nu} $ where $\nu$ is the relative angular motion, and $ 100 \text{km s}^{-1} $ is a typical value at $\text{4 kpc}$ and it gives $\nu \approx 1.7 \times 10^{-7} \text{mas s}^{-1} $.

It is estimated that for each star that there are 130 - 200 earth mass planets in the galaxy, and microlensing is the only method that can detect these planets.

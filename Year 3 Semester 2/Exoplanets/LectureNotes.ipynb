### The Two Body Problem
$$ \bold{F}_1 = m \ddot{\bold{r}}_1 = \frac{Gm_1m_2}{r^3} \bold{r}$$
$$ \bold{F}_2 = m \ddot{\bold{r}}_2 = -\frac{Gm_1m_2}{r^3} \bold{r}$$
Where $\bold{r} = \bold{r}_2 - \bold{r}_1$ and $\ddot{\bold{r}}=\ddot{\bold{r_2}}-\ddot{\bold{r_1}}$
$$\ddot{\bold{r}} - \frac{G m_1 \bold{r}}{r^3} + \frac{G m_2 \bold{r}}{r^3} = 0 $$
$$\ddot{\bold{r}} + G (m_1 + m_2) \frac{\bold{r}}{r^3}  = 0 $$
$$\ddot{\bold{r}} +\mu \frac{\bold{r}}{r^3} = 0$$

Taking the cross product of r with the two body equation gives:
$$ \bold{r} \times \ddot{\bold{r}} = -\mu \frac{\bold{r} \times \bold{r}}{r^3} = 0 $$
So angular momentum is conserved.
$$ \bold{r} \times \dot{\bold{r}} = \bold{h} $$
$$\bold{h} = \begin{vmatrix} \bold{r} \\\ 0 \\\ 0 \end{vmatrix} \times \begin{vmatrix} \dot{r} \\\ r \dot{\theta} \\\ 0 \end{vmatrix} = \begin{vmatrix} 0 \\\ 0 \\\ r^2 \dot{\theta} \end{vmatrix} = r^2 \dot{\theta} \hat{z}$$
by substitution of $r=u^{-1}$ we get:
$$ \frac{d^2 u}{d \theta^2} + u = \frac{\mu}{h^2} $$
The solution to this differential equation is:
$$ u(\theta) = \frac{\mu}{h^2} [1 + e \cos(\theta - \theta_0)] $$
Where e is the eccentricity and $\theta_0$ is the angle of periapsis.
#### Kepler's First Law
$$ r(\theta) = \frac{h^2 / \mu}{1 + e \cos(\theta - \theta_0)} $$
This describes a conic section with one focus at the origin.
#### Kepler's Second Law
$$ h = \sqrt{\mu a (1-e^2)} $$
$$ r = \frac{a(1-e^2)}{1 + e \cos(f)} $$
$$ dA = \int_{0}^{f} r dr d\theta $$
$$ \dot{A} = \frac{h}{2} = \frac{1}{2} \sqrt{\mu a (1-e^2)} = \text{constant} $$
#### Kepler's Third Law
$$ dA = \frac{h}{2} dT $$
$$ A = \frac{h}{2} T $$
$$ T = \frac{2A}{h} = \frac{2 \pi a^2 \sqrt{1-e^2}}{\sqrt{\mu a (1-e^2)}} $$
$$ T^2 = \frac{4 \pi^2}{\mu} a^3 = \frac{4 \pi^2}{G (m_1 + m_2)} a^3 $$
#### Vis-Viva Equation
$$ \frac{1}{2} v^2 - \frac{G (m_1 + m_2)}{a} = -\frac{G (m_1 + m_2)}{2a} $$
$$ v^2 = \frac{GM_1}{a} \frac{1+e}{1-e} $$

# Detection Methods
### Radial Velocity Method
#### The Doppler Effect
$$ \Delta \lambda = \lambda_{obs} - \lambda_{0} $$
$$ \lambda_{obs} = \lambda_{0} \frac{1 + v/c}{\sqrt{1 - \frac{v^2}{c^2}}} $$
which for $v \ll c$ reduces to:
$$ v=\frac{\Delta \lambda}{\lambda_0} c =V_{rad} = v \sin i $$
Where i is the inclination of the planet's orbit to the line of sight (i=90 degrees is edge on and i=0 degrees is face on).

#### Moving Around the Barycenter
$$ a = a_1 + a_2 $$
Where a is the distance between the star and the planet, a1 is the distance from the star to the barycenter and a2 is the distance from the planet to the barycenter.
$$ T_1 = T_2 = T $$
$$ e_1 = e_2 = e $$
$$ a_1 : a_2 : a = m_2 : m_1 : M $$
Where $ (M = m_1 + m_2) $
Kepler's Third Law gives:
$$ \frac{4 \pi^2}{G T^2} = \frac{M}{a^3} = \frac{m_2^3 / M^2}{a_1^3} = \frac{m_1^3 / M^2}{a_2^3} $$

#### Microlensing
Light deflection angle: $ \epsilon = \frac{4GM_L}{r c^2} $

Lens Equation: $ \theta_1^2 - \alpha \theta_1 - \theta_E^2 = 0 $

With Einstein angle: $ \theta_E = \sqrt{\frac{4GM_L}{\mu D_L c^2}} = \sqrt{\frac{2R_{\text{Sch}}}{\mu D_L}}$ ($\mu = 1+\frac{D_{L}}{D_SL}$)

$ u = \frac{\alpha}{\theta_E} $
$$ \frac{\theta_1^2}{\theta_E^2} - \frac{\alpha \theta_1}{\theta_E^2} - \frac{\theta_E^2}{\theta_E^2} = 0 $$

$y = \frac{\theta_1}{\theta_E}$
$$ y^2 - u y - 1 = 0 $$
$$ y = \frac{u \pm \sqrt{u^2 + 4}}{2} $$
$$ \theta_\pm = \frac{\theta_E}{2} (u \pm \sqrt{u^2 + 4}) $$
Magnification: $ A = \frac{u^2 + 2}{u \sqrt{u^2 + 4}} $

The timescale, $ t_E = \frac{\theta_E}{\nu} $ where $\nu$ is the relative angular motion, and $ 100 \text{km s}^{-1} $ is a typical value at $\text{4 kpc}$ and it gives $\nu \approx 1.7 \times 10^{-7} \text{mas s}^{-1} $.

It is estimated that for each star that there are 130 - 200 earth mass planets in the galaxy, and microlensing is the only method that can detect these planets.

### Atomospheres
Transit probability: $ P_{\text{transit}} = \frac{R_* + R_p}{a} \approx \frac{R_*}{a} $

Transit depth: $ D = \frac{F_{\text{out}}-F_{\text{in}}}{F_{\text{out}}} \approx \left( \frac{R_p}{R_*} \right)^2 $

Transit spectroscopy can be used to detect the atmosphere of a planet by looking at the absorption lines in the spectrum of the star as the planet transits in front of it. The depth of the transit will be different at different wavelengths due to the absorption of light by the atmosphere of the planet.

We can typically only view the upper part of the atmosphere down to 0.1 bar as the atmosphere becomes opaque at higher pressures.

### Atmospheres - Occultation Spectroscopy
Occultation spectroscopy is the study of the spectrum of a planet as it passes behind its host star. This can be used to detect the thermal emission from the planet and to study the composition of its atmosphere.

Looking at slide 285, the reason that the peaks after the transit (point 3) is not $F_\star$ is that the planet is still emitting a small amount of thermal radiation which contributes to the total flux observed.

Emmited FLux: $ B_\lambda (T) = \frac{2hc^2}{\lambda^5} \frac{1}{e^{\frac{hc}{\lambda kT}} - 1} $

in the Raleygh-Jeans limit ($\lambda \gg hc/kT$) this reduces to:
$$ B_\lambda (T) \approx \frac{2 k T c}{\lambda^4} $$
$$ \Delta F_{\text{em}}(\lambda) = D \frac{T_p}{T_\star} $$

Reflected flux, depth: $ \Delta F_{\text{ref}}(\lambda) = A_\lambda \left( \frac{r_p}{a} \right)^2 $

Total depth: $ \Delta F_{\text{occ}}(\lambda) = \Delta F_{\text{em}}(\lambda) + \Delta F_{\text{ref}}(\lambda) $

$$ \Delta_{\text{em}} = \frac{F_{\text{out}} - F_{\text{in}}}{F_{\text{out}}} = \frac{F_p + F_\star - F_{\star}}{F_p + F_\star} $$
$$ = \frac{F_p}{F_\star + F_p} \approx \frac{F_p}{F_\star} $$
$$ = \frac{r_p^2 T_p}{R_\star^2 T_\star + r_p^2 T_p} \approx \frac{r_p^2}{R_\star^2} \frac{T_p}{T_\star} $$

### Atmospheres Continued
Fundamentals of detecteion sources
$$ \text{SNR} = \frac{A}{\sigma_i / \sqrt{N}} $$
Where A is the signal we seek, $\sigma_i$ is the mean error and N is the number of points. The noise can be reduced by taking more observations, but the signal cannot be increased.

To make a detection, we need to test how different the data we just obtained is from a situation where there is no signal (the null hypothesis).

When SNR = 3, we usually have a $3\sigma$ detection, which corresponds to a p-value of 0.00135, meaning that there is only a 0.135% chance that the data we obtained is due to random fluctuations in the noise rather than a real signal.

However beware as if we take 300 detections at $3\sigma$, we would expect 1 of them to be a false positive due to random fluctuations in the noise. This is known as the multiple comparisons problem, and it is important to take it into account when interpreting the results of a detection.

### Orbital Migration
#### Disc-Driven Migration Type I - for Neptune mass and under
$$ \tau_I = \frac{H/a}{m_p/M_\star} \frac{M_\star}{\Sigma a^2} \Omega_p^{-1} $$
Where $H$ is the scale height of the disk, $a$ is the semi-major axis of the planet's orbit, $m_p$ is the mass of the planet, $M_\star$ is the mass of the star, $\Sigma$ is the surface density of the disk and $\Omega_p$ is the angular velocity of the planet and $\tau_I$ is the timescale for orbital migration.

$ \tau_I \approx 10^5 \text{ years} $ for an Earth-mass planet at 1 AU but is much shorter for more massive planets, which is a problem for planet formation theories as it means that planets would migrate into their host star before they have a chance to form.

#### Disc-Driven Migration Type II - for Jupiter mass and above
When a planet is massive enough to open a gap in the protoplanetary disk, it can undergo Type II migration. In this case, the planet is locked to the viscous evolution of the disk and migrates inward at the same rate as the disk material.
$$ \tau_{II} \approx \frac{a^2}{\alpha H^2 \Omega} \times \text{max}\left(1, \frac{m_p}{4\pi \Sigma a^2}\right) $$
Where $\alpha$ is the viscosity parameter of the disk, and the second term accounts for the fact that if the planet is more massive than the local disk mass, it will slow down the migration.

$ \tau_{II} \approx 10^5 \text{ years} $ for a Jupiter-mass planet at 1 AU, which is similar to the timescale for Type I migration, but it can be longer if the planet is more massive than the local disk mass.

### Grand-Tack Hypothesis
The Grand-Tack hypothesis is a model of the early solar system that explains the current configuration of the planets. It proposes that Jupiter formed at around 3.5 AU and then migrated inward to around 1.5 AU before reversing course and migrating back out to its current position at 5.2 AU. This migration would have caused a significant disturbance in the protoplanetary disk, leading to the formation of the asteroid belt and the current distribution of the terrestrial planets.

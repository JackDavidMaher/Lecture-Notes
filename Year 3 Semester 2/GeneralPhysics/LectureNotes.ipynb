```python
import numpy as np
import matplotlib.pyplot as plt

n = 1.5
wavelength = 700e-9

t = np.linspace(0, 500e-9, 10000)

# Phase difference including œÄ phase shift on top reflection
delta = 4 * np.pi * n * t / wavelength + np.pi

I = np.cos(delta / 2) ** 2

plt.figure(figsize=(6,3))
plt.plot(t * 1e9, I, lw=1.5)
plt.axvline(wavelength / (4 * n) * 1e9, color='red', ls='--', lw=1, label=f"t_min = {wavelength/(4*n)*1e9:.1f} nm")
plt.xlabel('Thickness (nm)')
plt.ylabel('Relative reflected intensity')
plt.title(f'Reflectance vs oil film thickness (Œª = {wavelength*1e9:.0f} nm)')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

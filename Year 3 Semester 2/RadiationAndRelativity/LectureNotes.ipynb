### Newtonian Mechanics

Newtons Laws:
1. A particle moves at a constant velocity (when resultant force = 0)
2. Relationship between acceleration and external force is $F = m \frac{d^2r}{dt^2} = ma$
3. Forces of action and reaction are equal magnitude and opposite direction

Galilean transformations:
1.  $r'=r-vt$
2.  $t'=t$

This means that the laws of motion are the same in all inertial frames.

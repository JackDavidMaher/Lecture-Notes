## 3 problems of the universe:
1. The flatness problem
    - Why is the universe so close to being spatially flat?
    - if omega (Ω) deviates slightly from 1 in the early universe, it would have diverged significantly over time, leading to a highly curved universe today. Yet, observations show that Ω is very close to 1.
2. The horizon problem
    - Why is the cosmic microwave background (CMB) so uniform in temperature across the sky, despite regions being out of causal contact?
3. The monopole problem
    - Why don't we observe magnetic monopoles, which are predicted by certain grand unified theories?
## Inflation Theory
d^2a/dt^2 / a = - (4πG/3) (ρ + 3pw)
if w < -1/3, then d^2a/dt^2 > 0 (accelerated expansion)
- Inflation proposes a period of rapid exponential expansion in the early universe, driven by a high-energy vacuum state (false vacuum) with negative pressure (w < -1/3).
- This rapid expansion can solve the flatness problem by driving Ω towards 1, the horizon problem by allowing regions to come into causal contact before inflation, and the monopole problem by diluting the density of monopoles to negligible levels.

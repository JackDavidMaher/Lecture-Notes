## Lecture 3
#### 3 problems of the universe:
1. The flatness problem
    - Why is the universe so close to being spatially flat?
    - if omega (Ω) deviates slightly from 1 in the early universe, it would have diverged significantly over time, leading to a highly curved universe today. Yet, observations show that Ω is very close to 1.
2. The horizon problem
    - Why is the cosmic microwave background (CMB) so uniform in temperature across the sky, despite regions being out of causal contact?
3. The monopole problem
    - Why don't we observe magnetic monopoles, which are predicted by certain grand unified theories?
#### Inflation Theory
d^2a/dt^2 / a = - (4πG/3) (ρ + 3pw)
if w < -1/3, then d^2a/dt^2 > 0 (accelerated expansion)
- Inflation proposes a period of rapid exponential expansion in the early universe, driven by a high-energy vacuum state (false vacuum) with negative pressure (w < -1/3).
- This rapid expansion can solve the flatness problem by driving Ω towards 1, the horizon problem by allowing regions to come into causal contact before inflation, and the monopole problem by diluting the density of monopoles to negligible levels.

## Lecture 4: Overdensity and Gravitational Instability
δ(**r**,t) <-- overdensity at position **r** and time t

δ(**r**,t) <-- Negative if underdense; minimum value -1 (voids)

δ(**r**,t) <-- positive if overdense; no maximum value (clusters)

δ(**r**,t) << 1 (linear regime)
- In the early universe, density fluctuations were small (δ << 1), allowing us to use linear perturbation theory to describe their growth.
- The evolution of these fluctuations can be described by the linearized fluid equations, leading to the growth of structures over time.
- As the universe expands, these small perturbations grow under the influence of gravity, eventually leading to the formation of galaxies and larger structures.
- The growth of perturbations is influenced by the matter content of the universe, with dark matter playing a crucial role in structure formation.

δ(**r**,t) ~ 1 (non-linear growth)
- As density fluctuations grow, they eventually reach a point where δ ~ 1, marking the transition to the non-linear regime.
- In this regime, the simple linear theory breaks down, and more complex processes such as gravitational collapse, mergers, and feedback mechanisms become important.
- Non-linear growth leads to the formation of bound structures like galaxies, clusters, and superclusters.
- Numerical simulations are often used to study the non-linear evolution of structures in the universe, providing insights into the complex dynamics involved.

δ(**r**,t) >> 1 (virialization)
- When density fluctuations become very large (δ >> 1), structures reach a state of virial equilibrium.
- Virialization occurs when the kinetic energy of a system balances its gravitational potential energy, leading to stable, bound structures.
- This process is crucial for the formation of galaxies and clusters, as it allows them to maintain their integrity against further collapse.
- The virial theorem provides a useful framework for understanding the dynamics of these systems in equilibrium.

$$ \ddot \delta = 4 \pi G \bar \rho \delta $$

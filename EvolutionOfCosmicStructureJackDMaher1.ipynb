### Question 1:
In Planck 2018 results. VI. Cosmological parameters from Astronomy & Astrophysics  August 10, 2021, the value of $\sigma_8$ is given as: $ \sigma_8 = 0.811\pm0.006$
In this value they assumed the base $ \Lambda \text{CDM}$ model. 

**Model dependence and limitations of the quoted $\sigma_8$**

- **Assumes base $\Lambda$CDM:** The Planck value is derived assuming flat $\Lambda$CDM with a standard set of parameters (e.g. fixed parameterisation of dark energy, a chosen neutrino model). If you relax these assumptions (allow curvature, evolving dark energy equation-of-state `w(z)`, or modified gravity), the mapping from the primordial amplitude to the present-day clustering amplitude changes and the posterior for $\sigma_8$ shifts.

- **Parameter degeneracies & priors:** CMB constraints restrict parameter combinations; for example $A_s e^{-2\tau}$ is well measured while $A_s$ and the optical depth $\tau$ are degenerate. Uncertainty or different priors on $\tau$, the sum of neutrino masses $\Sigma m_\nu$, or $n_s$ propagate into uncertainty and bias in $\sigma_8$.

- **High-z → low-z extrapolation:** The CMB measures fluctuations at $z\sim1100$. Converting to $\sigma_8$ at $z=0$ uses the assumed linear growth history and a transfer function. Physics that induces scale-dependent growth (massive neutrinos, warm dark matter, or modified gravity) makes a single-number $\sigma_8$ an incomplete description.

- **Nonlinear/baryonic corrections:** The 8 $h^{-1}$Mpc scale is partially nonlinear today and baryonic processes (AGN feedback, cooling) alter the matter power spectrum. Analyses use nonlinear prescriptions (e.g. Halofit or emulators) and imperfect modelling of baryons adds systematic uncertainty when relating the linear-theory amplitude to observables.

**Two short examples (illustrative):**

- *Massive neutrinos:* Allowing a nonzero $\Sigma m_\nu$ reduces small-scale growth because neutrinos free-stream out of perturbations. Typical cosmological chains show that increasing $\Sigma m_\nu$ from the minimal value to O(0.1–0.2 eV) can lower the inferred $\sigma_8$ by of order a few percent (roughly $\Delta\sigma_8\sim0.01$–0.03), depending on degeneracies with $\Omega_m$ and the priors used.

- *Varying dark-energy equation-of-state:* Letting $w\neq-1$ changes the growth rate at late times. Allowing $w$ to vary by ~0.1 typically alters $\sigma_8$ at the percent-level (order $10^{-2}$), though the exact shift depends on the correlated change in $\Omega_m$ and other parameters.

**Implication:** Because the Planck number is model-dependent and compresses scale/time information into one scalar, careful comparisons between CMB-derived $\sigma_8$ and low-redshift probes (weak lensing, clusters) should either (a) compare full parameter posteriors in the same model space, or (b) use more informative combinations like $S_8=\sigma_8(\Omega_m/0.3)^{0.5}$ or the full power-spectrum shape to avoid misinterpreting apparent tensions as new physics when they might arise from model choices or systematics.

**Reference (measurement cited above):** Planck Collaboration (N. Aghanim et al.), Planck 2018 results. VI. Cosmological parameters, A&A 641, A6 (2020). arXiv:1807.06209.

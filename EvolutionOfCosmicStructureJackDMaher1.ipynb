### Question 1:
In Planck 2018 results. VI. Cosmological parameters from Astronomy & Astrophysics  August 10, 2021, the value of $\sigma_8$ is given as: $ \sigma_8 = 0.811\pm0.006$
In this value they assumed the base $ \Lambda \text{CDM}$ model. 

Limitations:

- The Planck value assumes base $\Lambda \textrm{CDM}$. If you allow for curvature, the value of $\sigma_8$ shifts.

- High-z -> low-z extrapolation: The CMB measures fluctuations at $z\sim1100$. Converting to $\sigma_8$ at $z=0$ assumes linear growth. If there are non-linear effects this extrapolation could be inaccurate.


### Question 2:
$$ ds^2 = -c^2 dt^2 + a(t)^2 dr^2 $$
For light, $ ds^2 = 0 $ so $ dt = \frac{a(t)}{c} dr $ and the comoving distance is:
$$ \int_{r_e}^{r_0} dr = \int_{t_e}^{t_0} \frac{c \, dt}{a(t)} $$
$$ 0=\int_{t_e}^{t_e + \delta t_e} \frac{c \, dt}{a(t)} - \int_{t_0}^{t_0 + \delta t_0} \frac{c \, dt}{a(t)} $$
$$ \frac{c \, \delta t_e}{a(t_e)} = \frac{c \, \delta t_0}{a(t_0)} $$
$$ \frac{\delta t_0}{\delta t_e} = \frac{a(t_0)}{a(t_e)} = \frac{\nu_e}{\nu_0} = \frac{\lambda_0}{\lambda_e} = 1 + z $$
$$ 1 + z = \frac{a(t_0)}{a(t_e)} $$
if $ a(t_0) = 1 $ then $ a(t_e) = \frac{1}{1+z} $

### Question 3:
Equation of state parameter $ w = \frac{P}{\rho} = \frac{\dot \phi ^2 - V}{\dot \phi ^2 + V} $

Fluid continuity equation: $ \dot \rho + 3H(\rho + P) = 0 $

If field is constant, $ \dot \rho = 0 $ so $ \rho + P = 0 $ so $P = - \rho $ so $ w = -1 $

This means that the condition for a constant field is $ \dot \phi^2 \ll V(\phi) $ so that $ w \approx -1 $
